```python
import pandas as pd
```

```python
mps = pd.read_csv('data/current_mps.csv', sep=';')
mps[['nom', 'parti_ratt_financier']] .head()
```

```python
mps.info()
```

```python
print(len(mps[mps.sexe == 'H']))
print(len(mps[mps.sexe == 'F']))
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
labels = ['Hommes', 'Femmes']
sizes = [len(mps[mps.sexe == 'H']),
                 len(mps[mps.sexe == 'F'])]
colors = ['lightskyblue', 'lightcoral']
explode = (0, 0.1)
ax.pie(sizes, labels=labels, explode=explode, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')
ax.set_title(f"Global ({len(mps)} MPs)")
plt.savefig('parite.png')
plt.show()
```
